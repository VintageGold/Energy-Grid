```python
import pandas as pd
```

```python
df_value = pd.read_csv('6-12-2020.csv')
df_perct = pd.read_csv('6-12-2020_perct.csv')
df_inter = pd.read_csv('6-12-2020_inter.csv')
```

```python
df_perct.sample(5)
```

```python
value = df_value.merge(df_perct,on= list(df_value.columns)[:5])
```

```python
df = value.merge(df_inter,on=['Region Code','Region Type'])
```

```python
df = df.dropna(axis='columns',how='all')
```
